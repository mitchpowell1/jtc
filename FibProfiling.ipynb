```python
# Imports a couple of libraries that will help us visualize how our algorithms are behaving
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation

from time import perf_counter
```

```python
def collect_execution_times(f, n):
    out = []
    for i in range(n):
        start = perf_counter()
        f(i)
        end = perf_counter()
        out.append(end - start)
    return out
```

```python
# Define our efficient fibonacci function
def efficient_fibonacci(n):
    a = 0
    b = 1
    for i in range(n):
        tmp = a + b
        a = b
        b = tmp
    return a
```

```python
# Define our inefficient fibonacci function
def slow_fibonacci(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return slow_fibonacci(n - 2) + slow_fibonacci(n - 1)
```

```python
# Profile our functions
fast_fib_execution_times = collect_execution_times(efficient_fibonacci, 100)
slow_fib_execution_times = collect_execution_times(slow_fibonacci, 45)
```

```python
# Let's visualize how long it takes to execute as `n` grows
fig, ax = plt.subplots()
ns = [n for n in range(0, 100, 3)]
ts = [fast_fib_execution_times[n] for n in range(0, 100, 3)]
ax.set_title("Efficient Fibonacci Execution Times")
ax.set_ylabel("Execution Time (Seconds)")
ax.set_ylim([0, 0.00001])
ax.set_xlabel("n")
ax.set_xlim([0, 100])

scatter = ax.scatter(ns, ts)

plt.show()
```

```python
# Now lets look at our slow fibonacci implementation
fig, ax = plt.subplots()
ns = [n for n in range(45)]
ax.set_title("Inefficient Fibonacci Execution Times")
ax.set_ylabel("Execution Time (Seconds)")
ax.set_ylim([0, 300])
ax.set_xlabel("n")
ax.set_xlim([0, 45])

scatter = ax.scatter(ns, slow_fib_execution_times)

plt.show()
```
